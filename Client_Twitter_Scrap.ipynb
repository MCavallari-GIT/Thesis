```python
import twitter_scrap

timelines=["guardian","MailOnline","Telegraph","DailyMirror"]
start_time='2022-12-14T00:00:00.000Z'
delta=24*2  
# The Sun, The Times, The Telegraph, The Guardian, Daily Mail Online, Mirror, Independent, Sporting Life, Racing Post e BBC
```

## **Example : return all the tweets in a time window (start_date+delta)**

```python
twitter_scrap.retrive_tweets(clean_bearer2(bearer_file),timelines,mr=250,start_time=start_time,delta=delta,filename="test.csv")
```

```python
df=pd.read_csv("test.csv")
df
```
