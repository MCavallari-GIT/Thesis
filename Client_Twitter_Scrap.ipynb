```python
bearer_file = files.upload()
```

```python
def clean_bearer2(file):
  return str(list(file.values())[0])[2:len(str(list(file.values())[0]))-1]
```

```python
import twitter_scrap

timelines=["nytimes","guardian"]
start_time='2021-01-01T00:00:00.000Z'
delta=24*3  
```

## **Example 1: return ALL the tweets in a time window (start_date+delta)**

Asking all the tweets (*limit_fixed*=False) in 3 days (24x3) from te start date fo nytimes and the guardian. returned 100 tweets per time (*mr*=100). In this case, since *limit_fixed*=False, the parameter *tweets_per_timeline* is usless.

```python
twitter_scrap.retrive_many(clean_bearer2(bearer_file),timelines,tweets_per_timeline=50,mr=100,start_time=start_time,delta=delta,filename="example1.csv",limit_fixed=False)
```

```python
df=twitter_scrap.read_csv("example1.csv")
df
```

## **Example 2: return *tweets_per_timeline* tweets in a time window (start_date+delta)**

Asking a specific number of tweets (limit_fixed=True) in 3 days (24x3) from te start date for nytimes and the guardian. returned 100 tweets per time (mr=100). 

Please, insert a *tweets_per_timeline* as a multiple of *mr*.

```python
twitter_scrap.retrive_many(clean_bearer2(bearer_file),timelines,tweets_per_timeline=150,mr=50,start_time=start_time,delta=delta,filename="example2.csv",limit_fixed=True)
```
