```python
def clean_bearer2(file):
  return str(list(file.values())[0])[2:len(str(list(file.values())[0]))-1]
```

```python
import twitter_scrap

timelines=["guardian","MailOnline","Telegraph","DailyMirror"]
start_time='2022-12-14T00:00:00.000Z'
delta=24*2  
# The Sun, The Times, The Telegraph, The Guardian, Daily Mail Online, Mirror, Independent, Sporting Life, Racing Post e BBC
```

## **Example 1: return ALL the tweets in a time window (start_date+delta)**

Asking all the tweets (*limit_fixed*=False) in 3 days (24x3) from te start date fo nytimes and the guardian. returned 100 tweets per time (*mr*=100). In this case, since *limit_fixed*=False, the parameter *tweets_per_timeline* is usless.

```python
twitter_scrap.retrive_many(clean_bearer2(bearer_file),timelines,tweets_per_timeline=50,mr=500,start_time=start_time,delta=delta,filename="test.csv",limit_fixed=False)
```

```python
df=twitter_scrap.read_csv("test.csv")
df
```

```python
df['authorID'].unique()
```

```python
query="Defending world champions end Morocco’s World Cup dreams and book their place in the final. France have beaten Morocco 2-0 at Al Bayt Stadium to claim a place in Sunday’s World Cup final."
results=twitter_scrap.similarity(query,df['Text'].tolist())
results[:10]
```

```python
list(filter(lambda x: x[1] >= 0.5, results))
```
